```python
import random

class TTT(object):
    winning_combos = (
        [0, 1, 2], [3, 4, 5], [6, 7, 8],
        [0, 3, 6], [1, 4, 7], [2, 5, 8],
        [0, 4, 8], [2, 4, 6])

    winners = ('X wins', 'O win', 'Draw')

    def __init__(self):
        self.squares = [None for i in range(9)]
        

    def show(self):
        for element in [self.squares[i:i + 3] for i in range(0, len(self.squares), 3)]:
            print element
        print '\n'

    def available_moves(self):
        return [k for k, v in enumerate(self.squares) if v is None]

    def available_combos(self, player):
        return self.available_moves() + self.get_squares(player)

    def isOver(self):
        if None not in [v for v in self.squares]:
            return True
        if self.winner() != None:
            return True
        return False

    def winner(self):
        for player in ('X', 'O'):
            positions = self.get_squares(player)
            for combo in self.winning_combos:
                win = True
                for pos in combo:
                    if pos not in positions:
                        win = False
                if win:
                    return player
        return None
    
    #taken spaces
    def get_squares(self, player):
        return [k for k, v in enumerate(self.squares) if v == player]

    def make_move(self, position, player):
        self.squares[position] = player
        
    def minimax(self, node, player):
        if node.isOver():
            if self.winner() == 'X':
                return -1
            elif self.isOver() == True and self.winner() is None:
                return 0
            elif self.winner() == 'O':
                return 1
            best = None

        for move in node.available_moves():
            node.make_move(move, player)
            val = self.minimax(node, get_enemy(player))
            node.make_move(move, None)
            best = type(val)
            
            if player == 'O':
                if val > best:
                    best = val
            else:
                if val < best:
                    best = val
            return best


def checkMoves(board, player):
    a = -2
    choices = []
    if len(board.available_moves()) == 9:
        return 4
    for move in board.available_moves():
        board.make_move(move, player)
        #val = board.alphabeta(board, get_enemy(player), -2, 2)
        val = board.minimax(board, get_enemy(player))
        board.make_move(move, None)
        
        if val > a:
            a = val
            choices = [move]
        elif val == a:
            choices.append(move)
    return random.choice(choices)

#switch players
def get_enemy(player):
    if player == 'X':
        return 'O'
    return 'X'

def start():
    board = TTT()

    while not board.isOver():
        player = 'X'
        board.make_move(checkMoves(board, player), player)
        board.show()

        if board.isOver():
            break
        player = get_enemy(player)
        computer_move = checkMoves(board, player)
        board.make_move(computer_move, player)
        board.show()

    if board.winner() == 'X':
        print board.winners[0]
    elif board.winner() == 'O':
        print board.winners[1]
    else:
        print board.winners[2]
        
start()
```
